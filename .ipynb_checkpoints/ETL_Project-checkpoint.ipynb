```python
import requests
import json
import pandas as pd
```

```python
url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1960/1979/usa"
response = requests.get(url).json()
usa_60_79_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1980/1999/usa"
response = requests.get(url).json()
usa_80_99_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1960/1979/chn"
response = requests.get(url).json()
chn_60_79_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1980/1999/chn"
response = requests.get(url).json()
chn_80_99_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1960/1979/bra"
response = requests.get(url).json()
bra_60_79_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1980/1999/bra"
response = requests.get(url).json()
bra_80_99_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1960/1979/deu"
response = requests.get(url).json()
deu_60_79_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1980/1999/deu"
response = requests.get(url).json()
deu_80_99_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1960/1979/ind"
response = requests.get(url).json()
ind_60_79_temp = response[0]["monthVals"]

url = "http://climatedataapi.worldbank.org/climateweb/rest/v1/country/mavg/tas/1980/1999/ind"
response = requests.get(url).json()
ind_80_99_temp = response[0]["monthVals"]

# print(json.dumps(response, indent=4, sort_keys=True))
```

```python
# response[0]["monthVals"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temp_dictionary = {
    "Months": months,
    "USA 1960-1979": usa_60_79_temp,
    "USA 1980-1999": usa_80_99_temp,
    "Brazil 1960-1979": bra_60_79_temp,
    "Brazil 1980-1999": bra_80_99_temp,
    "China 1960-1979": chn_60_79_temp,
    "China 1980-1999": chn_80_99_temp,
    "Germany 1960-1979": deu_60_79_temp,
    "Germany 1980-1999": deu_80_99_temp,
    "India 1960-1979": ind_60_79_temp,
    "India 1980-1999": ind_80_99_temp,
    
}
temp_dictionary_pd = pd.DataFrame(temp_dictionary)
temp_dictionary_pd
```

```python
# Find the average temperature for 1980-1999 by taking the mean of all the months

USA_1980_1999_avtemp = temp_dictionary_pd.loc[:,"USA 1980-1999"].mean()

BRA_1980_1999_avtemp = temp_dictionary_pd.loc[:,"Brazil 1980-1999"].mean()

CHN_1980_1999_avtemp = temp_dictionary_pd.loc[:,"China 1980-1999"].mean()

DEU_1980_1999_avtemp = temp_dictionary_pd.loc[:,"Germany 1980-1999"].mean()

IND_1980_1999_avtemp = temp_dictionary_pd.loc[:,"India 1980-1999"].mean()



# Create a new data frame with all the means
temp_averages = {
    "USA 1980-1999": USA_1980_1999_avtemp,
    "Brazil 1980-1999": BRA_1980_1999_avtemp,
    "China 1980-1999": CHN_1980_1999_avtemp,
    "Germany 1980-1999": DEU_1980_1999_avtemp,
    "India 1980-1999": IND_1980_1999_avtemp
    
}

temp_averages_df = pd.DataFrame([temp_averages])
temp_averages_df
```

```python
path_1 = "csv/green_house_gas.csv"
green_house_gas = pd.read_csv(path_1)
green_house_gas_df = pd.DataFrame(green_house_gas)
green_house_gas_df.head()
```

```python
# Drop unnecessary columns
green_house_gas_df = green_house_gas_df.drop(columns=['Country Code','Indicator Name','Indicator Code', 'Unnamed: 63'])
green_house_gas_df.head()
```

```python
# Select necessary rows for the 5 countries
green_house_gas_df = green_house_gas_df.loc[green_house_gas_df['Country Name'].isin(['United States','Brazil','China', 'Germany', 'India'])]
green_house_gas_df
```

```python
# Drop NaN values
green_house_gas_df = green_house_gas_df.dropna(axis='columns')
green_house_gas_df
```

```python
# Drop years that are not 1980-1999
green_house_gas_clean = green_house_gas_df.drop(columns=['1970','1971','1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012'])
green_house_gas_clean.head()
```
