```python
# response[0]["monthVals"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temp_dictionary = {
    "Months": months,
    "USA 1960-1979": usa_60_79_temp,
    "USA 1980-1999": usa_80_99_temp,
    "Brazil 1960-1979": bra_60_79_temp,
    "Brazil 1980-1999": bra_80_99_temp,
    "China 1960-1979": chn_60_79_temp,
    "China 1980-1999": chn_80_99_temp,
    "Germany 1960-1979": deu_60_79_temp,
    "Germany 1980-1999": deu_80_99_temp,
    "India 1960-1979": ind_60_79_temp,
    "India 1980-1999": ind_80_99_temp,
    
}
temp_dictionary_pd = pd.DataFrame(temp_dictionary)
temp_dictionary_pd
```

```python
# Find the average temperature for 1980-1999 by taking the mean of all the months

USA_1980_1999_avtemp = temp_dictionary_pd.loc[:,"USA 1980-1999"].mean()

BRA_1980_1999_avtemp = temp_dictionary_pd.loc[:,"Brazil 1980-1999"].mean()

CHN_1980_1999_avtemp = temp_dictionary_pd.loc[:,"China 1980-1999"].mean()

DEU_1980_1999_avtemp = temp_dictionary_pd.loc[:,"Germany 1980-1999"].mean()

IND_1980_1999_avtemp = temp_dictionary_pd.loc[:,"India 1980-1999"].mean()



# Create a new data frame with all the means
temp_averages = {
    "United States": USA_1980_1999_avtemp,
    "Brazil": BRA_1980_1999_avtemp,
    "China": CHN_1980_1999_avtemp,
    "Germany": DEU_1980_1999_avtemp,
    "India": IND_1980_1999_avtemp
    
}

temp_averages_df = pd.DataFrame([temp_averages])
temp_averages_df
```

```python
# Transpose Temperature Data
temp_averages_df = temp_averages_df.T.reset_index()
temp_averages_final = temp_averages_df.rename(columns={'index': 'country_name', 0: 'Temperature_1980_1999'})
temp_averages_final
```
